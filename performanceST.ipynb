# EXPLORING DATA


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df["mean score"] =((df["math score"]+df["reading score"]+df["writing score"])/3).round()
df.head()
```

```python
df['gender'].value_counts()
```

# label encoding

```python
from sklearn.preprocessing import LabelEncoder
lc = LabelEncoder()
df['gender'] = lc.fit_transform(df['gender'])
df['race/ethnicity'] = lc.fit_transform(df['race/ethnicity'])
df['lunch'] = lc.fit_transform(df['lunch'])
df['test preparation course'] = lc.fit_transform(df['test preparation course'])
df.head()
```

```python
level ={
    "bachelor's degree":0,
    "some college":1,
    "master's degree":2,
    "associate's degree":3,
    "high school":4,
    "some high school":5
}
```

```python
df['parental level of education']=df['parental level of education'].map(level)
```

```python
df = pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

```python
x = df.drop(columns = 'mean score').values
```

```python
x
```

```python
x[0]
```

```python
y = df['mean score'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
predictions= model.predict(x_test)
```

```python
predictions
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(predictions,y_test))
```
